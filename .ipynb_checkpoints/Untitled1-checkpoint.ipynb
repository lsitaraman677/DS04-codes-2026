```python
    import sys
    n_particles = 20
    iters= list(range(0, 400, 20))
    n_iters= len(iters)

    data1 = np.full((n_iters, n_particles, 3), np.nan)
    data2 = np.full((n_iters, n_particles, 3), np.nan)

    def arr_to_idxdict(arr):
        d = {}
        for i in range(len(arr)):
            d[arr[i]] = i
        return d
    
    beam1_ids = np.array(pIN.get_particle(species='beam1', var_list=['id'], iteration=0)[0])
    beam1_ids = arr_to_idxdict(beam1_ids[np.linspace(0, len(beam1_ids)-1, 20, dtype=np.int32)])

    beam2_ids = np.array(pIN.get_particle(species='beam2', var_list=['id'], iteration=0)[0])
    beam2_ids = arr_to_idxdict(beam2_ids[np.linspace(0, len(beam2_ids)-1, 20, dtype=np.int32)])

    for i in range(n_iters):
        beam1 = np.array(pIN.get_particle(species='beam1', var_list=['z', 'x', 'y', 'id'], iteration=iters[i])).T
        weee = [0 for _ in range(20)]
        for j in range(beam1.shape[0]):
            if beam1[j, 3] in beam1_ids:
                data1[i, beam1_ids[beam1[j, 3]]] = beam1[j, :3]
                weee[beam1_ids[beam1[j, 3]]] += 1
        beam2 = np.array(pIN.get_particle(species='beam2', var_list=['z', 'x', 'y', 'id'], iteration=iters[i])).T
        for j in range(beam2.shape[0]):
            if beam2[j, 3] in beam2_ids:
                data2[i, beam2_ids[beam2[j, 3]]] = beam2[j, :3]
                weee[beam2_ids[beam2[j, 3]]] += 1
        if 1 in weee or 0 in weee:
            print(weee)
        print(f'finished {i} iterations')
        plt.figure()
        plt.show()
```
